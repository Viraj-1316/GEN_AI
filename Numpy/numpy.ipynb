Task - 3


```python
values = np.array([2,4,6,8,10])

square_root = np.sqrt(values)
print(square_root)

exponential = np.exp(values)
print(exponential)

natural_log = np.log(values)
print(natural_log)

sum = np.sum(values)
print(sum)

"""
    Cumulative sum means adding elements of an array step 
    by step, keeping track of the total as you go.
"""
cumulative_sum = np.cumulative_sum(values)
print(cumulative_sum)
```

```python
data = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,90]])

row_wise_sum = np.sum(data, axis=1)
print(row_wise_sum)

col_wise_sum = np.sum(data, axis=0)
print(col_wise_sum)

min_value = np.min(data)
print(min_value)

max_value = np.max(data)
print(max_value)

mean = np.mean(data)
print(mean)
```

```python
sales = np.array([1200,1500,900,2000,1800,1700,1600])

print('total weekly sales =', np.sum(sales))

print('avg daily sales =', np.mean(sales))

print('highest and lowest sales day =', np.max(sales), np.min(sales))

print('standard deviation of sales =', np.std(sales))

avg = np.mean(sales)
print('sales above avg sales =', sales[sales>avg])
```
